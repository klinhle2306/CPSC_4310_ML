## Advertising data

Let's take a look at some data, ask some questions about that data, and then use linear regression to answer those questions!

A toy dataset for your exploratory data analysis:

./data/Advertising.csv

```python
import pandas as pd
import numpy
import statistics as st
from matplotlib import pyplot as plt

# read CSV file stored in the current directory and save the results
data = pd.read_csv('./data/Advertising.csv', skipinitialspace=True, index_col=0)
# display the first 5 rows
data.head()
```

What are the features?
- **TV:** advertising dollars spent on TV for a single product in a given market (in thousands of dollars)
- **Radio:** advertising dollars spent on Radio
- **Newspaper:** advertising dollars spent on Newspaper

What is the response?
- **Sales:** sales of a single product in a given market (in thousands of items)

What else do we know?
- Because the response variable is continuous, this is a **regression** problem.
- There are 200 **observations** (represented by the rows), and each observation is a single market.

```python
#1
TVCost = data['TV']
RadioCost = data['Radio']
NewspaperCost = data['Newspaper']
SaleTotal = data['Sales']

print("Mean: (TV, Radio, Newspaper, Sales)")
print(st.mean(TVCost))
print(st.mean(RadioCost))
print(st.mean(NewspaperCost))
print(st.mean(SaleTotal))
print()
print("Median: (TV, Radio, Newspaper, Sales)")
print(st.median(TVCost))
print(st.median(RadioCost))
print(st.median(NewspaperCost))
print(st.median(SaleTotal))
print()
print("Mode: (TV, Radio, Newspaper, Sales)")
print(st.mode(TVCost))
print(st.mode(RadioCost))
print(st.mode(NewspaperCost))
print(st.mode(SaleTotal))

```

```python
#2
print('TV')
print('min:', min(TVCost))
print('max:', max(TVCost))
print('median:', st.median(TVCost))
print('first quartile:', numpy.quantile(TVCost, 0.25))
print('third quartile:', numpy.quantile(TVCost, 0.75))

print()
print('Radio')
print('min:', min(RadioCost))
print('max:', max(RadioCost))
print('median:', st.median(RadioCost))
print('first quartile:', numpy.quantile(RadioCost, 0.25))
print('third quartile:', numpy.quantile(RadioCost, 0.75))

print()
print('Newspaper')
print('min:', min(NewspaperCost))
print('max:', max(NewspaperCost))
print('median:', st.median(NewspaperCost))
print('first quartile:', numpy.quantile(NewspaperCost, 0.25))
print('third quartile:', numpy.quantile(NewspaperCost, 0.75))

print()
print('Sales')
print('min:', min(SaleTotal))
print('max:', max(SaleTotal))
print('median:', st.median(SaleTotal))
print('first quartile:', numpy.quantile(SaleTotal, 0.25))
print('third quartile:', numpy.quantile(SaleTotal, 0.75))
```

```python
#3
plt.boxplot(data)
plt.show()
```

```python
#4
plt.hist(SaleTotal)
plt.show()
```

```python
#5
plt.figure(figsize=(10,10))
plt.plot(SaleTotal)
```

```python
#6
print('Pearson\'s correlation coefficient factor r:')
print('TV - Sales:')
print(numpy.corrcoef(TVCost, SaleTotal)[0, 1])
print('Radio - Sales:')
print(numpy.corrcoef(RadioCost, SaleTotal)[0, 1])
print('Newspaper - Sales:')
print(numpy.corrcoef(NewspaperCost, SaleTotal)[0, 1])
```

```python
#7
plt.scatter(TVCost, SaleTotal)
plt.title('TV-Sales')
plt.show()

plt.scatter(RadioCost, SaleTotal)
plt.title('Radio-Sales')
plt.show()

plt.scatter(NewspaperCost, SaleTotal)
plt.title('Newspaper-Sales')
plt.show()
```

```python
#8 Different graphs show us different information about the data. 
#For instance, box plot lets us know the general information about the minimum, the first quartile (Q1), median, the third quartile (Q3), and maximum of the data.
#Histogram tells us the frequency of data where scatter plot tells us the relationshp between data
```
